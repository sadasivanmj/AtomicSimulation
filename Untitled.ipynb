```python
import matplotlib.pyplot as plt
import numpy as np

Z = 1
e = 1.6e-19
r = np.linspace(0,1,1000)
l = 2
hbar = 1.05457182e-34
mu = 9e-31
eps0 = 8.85e-12

Veff = (-Z*e**2)/(4*np.pi*eps0*r) + (l*(l+1)*hbar**2)/(2*mu*r**2)
plt.plot(r,Veff)
```
