```python
import numpy as np
import pandas as pd

# Function to read and process large files safely
def read_file(filename, columns=None):
    data = []
    with open(filename, "r") as file:
        for line in file:
            try:
                values = line.split()
                data.append([float(value) for value in values])
            except ValueError:
                print(f"Skipping invalid line: {line.strip()}")
    return pd.DataFrame(data=data, columns=columns)

# Read PSdelta_frAr_NUM.out
df_delta = read_file("PSdelta_frAr_NUM.out", columns=["l", "E", "Delta"])
df_delta_s = df_delta[df_delta["l"] == 0]
df_delta_d = df_delta[df_delta["l"] == 2]

# Read sigma.in
df_sigma = read_file("sigma.in", columns=["E", "0", "DeltaL0", "2", "DeltaL2"])

# Read bouncontinuum.dat
df_matrix_element = read_file("bouncontinuum.dat", columns=["l", "E(j)", "matelement(j)"])
df_matrix_element_s = df_matrix_element[df_matrix_element["l"] == 0]
df_matrix_element_d = df_matrix_element[df_matrix_element["l"] == 2]

# Verify sizes of the dataframes
print("Sizes of DataFrames:")
print(f"Delta: {df_delta.size}, Sigma: {df_sigma.size}")
print(f"Matrix Element (S-wave): {df_matrix_element_s.size}, Matrix Element (D-wave): {df_matrix_element_d.size}")

# Ensure all arrays are aligned in size
min_size = min(len(df_sigma), len(df_delta_d), len(df_matrix_element_s), len(df_matrix_element_d))
df_sigma = df_sigma.iloc[:min_size]
df_delta_d = df_delta_d.iloc[:min_size]
df_delta_d.loc[df_delta_d.iloc[:, 2] >= 0, df_delta_d.columns[2]] -= np.pi
df_matrix_element_s = df_matrix_element_s.iloc[:min_size]
df_matrix_element_d = df_matrix_element_d.iloc[:min_size]

# Compute the complex phase shift
try:
    numerator = (
        np.sin(df_sigma["DeltaL0"].values + df_delta_s["Delta"].values) * df_matrix_element_s["matelement(j)"].values / 2
        - np.sin(df_sigma["DeltaL2"].values + df_delta_d["Delta"].values) * df_matrix_element_d["matelement(j)"].values
    )
    denominator = (
        np.cos(df_sigma["DeltaL0"].values + df_delta_s["Delta"].values) * df_matrix_element_s["matelement(j)"].values / 2
        - np.cos(df_sigma["DeltaL2"].values + df_delta_d["Delta"].values) * df_matrix_element_d["matelement(j)"].values
    )
    complex_phaseshift = np.arctan2(numerator, denominator)  # Use arctan2 for better numerical stability
except Exception as e:
    print(f"Error during complex phase shift computation: {e}")
    complex_phaseshift = None

# Save the results
if complex_phaseshift is not None:
    result_df = pd.DataFrame({
        "E": df_sigma["E"][:min_size],
        "Complex_PhaseShift": complex_phaseshift
    })
    result_df.to_csv("complex_phaseshift.out", index=False, sep="\t")
    print("Results saved to complex_phaseshift.out")

# Plot the results
import matplotlib.pyplot as plt

plt.plot( df_sigma["E"],complex_phaseshift, label="Complex Phase Shift")
plt.xlabel("Energy (E)")
plt.ylabel("Complex Phase Shift")
plt.title("Complex Phase Shift vs Energy")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def read_file_to_dataframe(file_path, columns):
    """Reads a file and returns a DataFrame with the specified columns."""
    try:
        with open(file_path, "r") as file:
            data = [
                [float(value) for value in line.split()]
                for line in file if line.strip()
            ]
        return pd.DataFrame(data, columns=columns)
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
        return pd.DataFrame()
    except ValueError:
        print(f"Error: File '{file_path}' contains invalid data.")
        return pd.DataFrame()

def filter_by_quantum_number(df, quantum_number):
    """Filters a DataFrame by the quantum number 'l'."""
    return df[df["l"] == quantum_number]

def plot_phase_shifts(df_s, df_d):
    """Plots phase shifts for s-wave and d-wave."""
    plt.figure(figsize=(8, 6))
    plt.plot(df_s["E"], df_s["Delta"], label="s-wave (l=0)", marker="o")
    plt.plot(df_d["E"], df_d["Delta"], label="d-wave (l=2)", marker="x")
    plt.xlabel("Energy (E)")
    plt.ylabel("Phase Shift (Delta)")
    plt.title("Phase Shifts vs Energy")
    plt.legend()
    plt.grid()
    plt.xlim(0,1)
    plt.savefig("partialphaseshift.png", format="png", dpi=300)

def plot_matrix_elements(df_s, df_d):
    """Plots matrix elements for s-wave and d-wave."""
    plt.figure(figsize=(8, 6))
    plt.plot(df_s["E(j)"], df_s["matelement(j)"], label="s-wave (l=0)", marker="o")
    plt.plot(df_d["E(j)"], df_d["matelement(j)"], label="d-wave (l=2)", marker="x")
    plt.xlabel("Energy (E)")
    plt.ylabel("Matrix Element")
    plt.title("Matrix Elements vs Energy")
    plt.legend()
    plt.grid()
    plt.show()

# Load and process delta data (phase shifts)
delta_file = "PSdelta_frAr_NUM.out"
df_delta = read_file_to_dataframe(delta_file, ["l", "E", "Delta"])
if not df_delta.empty:
    df_delta_s = filter_by_quantum_number(df_delta, 0)
    df_delta_d = filter_by_quantum_number(df_delta, 2)
    df_delta_d.loc[df_delta_d.iloc[:, 2] >= 0, df_delta_d.columns[2]] -= np.pi

# Load and process sigma data (cross-sections)
sigma_file = "sigma.in"
df_sigma = read_file_to_dataframe(sigma_file, ["E", "0", "DeltaL0", "2", "DeltaL2"])


# Load and process matrix element data
matrix_element_file = "bouncontinuum.dat"
df_matrix_element = read_file_to_dataframe(matrix_element_file, ["l", "E(j)", "matelement(j)"])
if not df_matrix_element.empty:
    df_matrix_element_s = filter_by_quantum_number(df_matrix_element, 0)
    df_matrix_element_d = filter_by_quantum_number(df_matrix_element, 2)

# Plot phase shifts
if not df_delta.empty:
    plot_phase_shifts(df_delta_s, df_delta_d)

# Plot matrix elements
if not df_matrix_element.empty:
    plot_matrix_elements(df_matrix_element_s, df_matrix_element_d)

if not df_sigma.empty:
    plt.figure(figsize=(8, 6))
    plt.plot(df_sigma["E"], df_sigma["DeltaL0"], label="s-wave (l=0)", marker="o")
    plt.plot(df_sigma["E"], df_sigma["DeltaL2"], label="d-wave (l=0)", marker="o")
    plt.xlabel("Energy (E)")
    plt.ylabel("Coulomb Phase Shift")
    plt.title("Phase Shift vs Energy")
    plt.xlim(0,1)
    plt.savefig("coulombphaseshift.png",format="png",dpi=300)
```
